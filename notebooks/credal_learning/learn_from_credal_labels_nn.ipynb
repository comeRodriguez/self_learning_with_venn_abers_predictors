# Import libraries

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.base import BaseEstimator

from typing import Tuple
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using {device} device")
```

# Load dataset

```python
def load_planar_dataset(n_example: int) -> Tuple[np.ndarray, np.ndarray]:
    """Create and return a mock dataset. If we plot the data, we can see
    a sort of flower. The created dataset contains n_example X of 2 features
    and their associated label

    Args:
        n_example (int): number of examples in the wanted dataset

    Returns:
        Tuple[np.ndarray, np.ndarray]: Features matrix and associated labels vector
    """
    np.random.seed(1)
    n_representation = int(n_example/2)
    n_features = 2
    features = np.zeros((n_example, n_features))
    labels = np.zeros((n_example,1), dtype='uint8')
    max_ray = 4
    for label in range(2):
        index = range(n_representation*label,n_representation*(label+1))
        theta = np.linspace(
            label*3.12,(label+1)*3.12,n_representation
        ) + np.random.randn(n_representation)*0.2
        radius = max_ray*np.sin(4*theta) + np.random.randn(n_representation)*0.2
        features[index] = np.c_[radius*np.sin(theta), radius*np.cos(theta)]
        labels[index] = label
    return features, labels
```

```python
df2 = pd.read_csv("../../data/SynthPara_n1000_p2.csv")
df2["y"] = df2["z"].apply(lambda x: 0 if x == "B" else 1)
df2.head()
```

```python
def transform_to_interval(variable: str) -> np.ndarray:
    """
    Transform a hard label into an interval of probabilities.
    If the label is 0, the created interval will be in the range of
    [0, 0.49], else it will be in the range of [0.5, 1]
    
    Args:
        variable (int): variable to transform
    
    Returns:
        (np.ndarray): created interval 
    """
    if variable == "B":
        interval = np.sort(
            np.round(
                np.random.uniform(
                    low=0.0,
                    high=0.49,
                    size=(2,)
                ),
                decimals=2
            )
        )
    else:
        interval = np.sort(
            np.round(
                np.random.uniform(
                    low=0.5,
                    high=1.0,
                    size=(2,)
                ),
                decimals=2
            )
        )
    return interval
```

```python
df2["interval"] = df2["z"].apply(transform_to_interval)
df2.head()
```

```python
def KLD_loss(y_true, y_pred):
    kld = nn.KLDivLoss(reduction="batchmean", log_target=False)
    loss = kld(y_pred, y_true) + kld(1-y_pred, 1-y_true)
    return loss

class NeuralNetwork(nn.Module):
    def __init__(self, loss_used):
        super(NeuralNetwork, self).__init__()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(2, 1),
            nn.Sigmoid(),
        )
        if loss_used == "binary_cross_entropy":
            self.loss = nn.BCELoss(reduction="mean")
        else:
            self.loss = None
        self.learning_rate = 0.0001
        self.optimizer = torch.optim.SGD(self.parameters(),lr=self.learning_rate)
        self.losses = []
        self.accuracies = []

    def fit(self, x_train, y_train, epochs=10):
        for _ in range(epochs):
            logits = self.linear_relu_stack(x_train)
            if self.loss:
                loss = self.loss(logits, y_train.reshape(-1, 1))
            else:
                loss = KLD_loss(y_train.reshape(-1, 1), logits)
            self.losses.append(loss.detach().numpy())
            self.accuracies.append(accuracy_score(
                y_true=y_train, y_pred=logits.reshape(-1).detach().numpy().round()))
            loss.backward()
            self.optimizer.step()
        return logits
```

```python
model = NeuralNetwork(loss_used="binary_cross_entropy").to(device)
print(model)
x_train = torch.from_numpy(df2[["X1", "X2"]].values.astype(np.float32))
y_train = torch.from_numpy(df2["y"].values.astype(np.float32))
logits = model.fit(x_train, y_train, epochs=200)
_, axes = plt.subplots(1, 2, figsize=(20, 5))
sns.lineplot(x=[i for i in range(200)], y=[float(loss) for loss in model.losses], ax=axes[0]).set(title="Loss")
sns.lineplot(x=[i for i in range(200)], y=model.accuracies, ax=axes[1]).set(title="Accuracy")
plt.show()
```

```python
def KLD_loss_intervals(y_true, y_pred):
    y_true = torch.from_numpy(np.stack(y_true.values).astype(np.float32))
    y_true_inf = y_true[:, 0].reshape(-1, 1)
    y_true_sup = y_true[:, 1].reshape(-1, 1)
    used_probas = torch.zeros(y_true.shape[0], 1)
    used_probas[y_pred <= y_true_inf] = y_true_inf[y_pred <= y_true_inf]
    used_probas[y_pred >= y_true_sup] = y_true_inf[y_pred >= y_true_sup]
    used_probas[(y_pred >= y_true_inf) & (y_pred <= y_true_sup)
                ] = y_pred[(y_pred >= y_true_inf) & (y_pred <= y_true_sup)]
    kld = nn.KLDivLoss(reduction="batchmean", log_target=False)
    loss = kld(y_pred, used_probas) + kld(1-y_pred, 1-used_probas)
    return loss
```

```python
class NeuralNetworkIntervals(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(2, 1),
            nn.Sigmoid(),
        )
        self.learning_rate = 0.1
        self.optimizer = torch.optim.SGD(self.parameters(),lr=self.learning_rate)
        self.losses = []
        self.accuracies = []

    def fit(self, x_train, y_train, hypotetic_labels, epochs=10):
        for _ in range(epochs):
            logits = self.linear_relu_stack(x_train)
            loss = KLD_loss_intervals(y_train, logits)
            self.losses.append(loss.detach().numpy())
            self.accuracies.append(accuracy_score(
                y_true=hypotetic_labels, y_pred=logits.reshape(-1).detach().numpy().round()))
            loss.backward()
            self.optimizer.step()
        return logits
```

```python
model = NeuralNetworkIntervals().to(device)
print(model)
x_train = torch.from_numpy(df2[["X1", "X2"]].values.astype(np.float32))
y_train = df2["interval"]
logits = model.fit(x_train, y_train, df2["y"], epochs=200)
_, axes = plt.subplots(1, 2, figsize=(20, 5))
sns.lineplot(x=[i for i in range(200)], y=[float(loss) for loss in model.losses], ax=axes[0]).set(title="Loss")
sns.lineplot(x=[i for i in range(200)], y=model.accuracies, ax=axes[1]).set(title="Accuracy")
plt.show()
```

```python
def KLD_loss(y_true, y_pred):
    kld = nn.KLDivLoss(reduction="batchmean", log_target=False)
    loss = kld(y_pred, y_true) + kld(1-y_pred, 1-y_true)
    return loss

class NeuralNetwork(nn.Module):
    def __init__(self, loss_used, learning_rate):
        super(NeuralNetwork, self).__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(2, 10),
            nn.Sigmoid(),
            nn.Linear(10, 1),
            nn.Sigmoid(),
        )
        if loss_used == "binary_cross_entropy":
            self.loss = nn.BCELoss(reduction="mean")
        else:
            self.loss = None
        self.learning_rate = learning_rate
        self.optimizer = torch.optim.SGD(self.parameters(),lr=self.learning_rate)
        self.losses = []
        self.accuracies = []

    def fit(self, x_train, y_train, epochs=10):
        for _ in range(epochs):
            logits = self.linear_relu_stack(x_train)
            if self.loss:
                loss = self.loss(logits, y_train.reshape(-1, 1))
            else:
                loss = KLD_loss(y_train.reshape(-1, 1), logits)
            self.losses.append(loss.detach().numpy())
            self.accuracies.append(accuracy_score(
                y_true=y_train, y_pred=logits.reshape(-1).detach().numpy().round()))
            loss.backward()
            self.optimizer.step()
        return logits
```
