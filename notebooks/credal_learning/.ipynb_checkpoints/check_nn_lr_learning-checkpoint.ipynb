```python
def transform_to_interval_int(variable: int) -> np.ndarray:
    """
    Transform a hard label into an interval of probabilities.
    If the label is 0, the created interval will be in the range of
    [0, 0.49], else it will be in the range of [0.5, 1]
    
    Args:
        variable (int): variable to transform
    
    Returns:
        (np.ndarray): created interval 
    """
    if variable == 0:
        interval = np.sort(
            np.round(
                np.random.uniform(
                    low=0.0,
                    high=0.49,
                    size=(2,)
                ),
                decimals=2
            )
        )
    else:
        interval = np.sort(
            np.round(
                np.random.uniform(
                    low=0.5,
                    high=1.0,
                    size=(2,)
                ),
                decimals=2
            )
        )
    return interval
```

```python
data["target_intervals"] = data["target"].apply(transform_to_interval_int)
data.head()
```

# With classical logistic regression using PyTorch 

```python
class LogisticRegressionPytorch(torch.nn.Module):
    def __init__(self, input_dim, output_dim, learning_rate):
        super().__init__()
        self.logistic = nn.Sequential(
            nn.Linear(input_dim, output_dim),
            nn.Sigmoid()
        )
        self.criterion = torch.nn.BCELoss(reduction="mean")
        self.optimizer = torch.optim.SGD(self.parameters(), lr=learning_rate)
        self.losses = []
        self.accuracies = []
        
    def fit(self, x_train, y_train, epochs=10):
        for _ in range(epochs):
            outputs = self.logistic(x_train)
            loss = self.loss(outputs, y_train.reshape(-1, 1))
            self.losses.append(loss.detach().numpy())
            self.accuracies.append(accuracy_score(
                y_true=y_train, y_pred=logits.reshape(-1).detach().numpy().round()))
        outputs = torch.sigmoid(self.linear(x))
        return outputs
```

```python
epochs = 200
input_dim = 30 
output_dim = 1 # Single binary output 
learning_rate = 0.01

X_train = torch.from_numpy(data.drop(["target", "target_intervals"], axis=1).values.astype(np.float32))
Y_train = torch.from_numpy(data["target"].values.astype(np.float32))

losses = []
accuracies = []

model = LogisticRegressionPytorch(input_dim,output_dim)
criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
print(model)
for epoch in range(epochs):
    x = X_train
    labels = Y_train
    outputs = model.forward(X_train)
    loss = criterion(torch.squeeze(outputs), labels) 
    loss.backward() # Computes the gradient of the given tensor w.r.t. the weights/bias
    
    optimizer.step() # Updates weights and biases with the optimizer (SGD)
    
    with torch.no_grad():
        # Calculating the loss and accuracy for the train dataset
        total = 0
        correct = 0
        total += Y_train.size(0)
        correct += np.sum(torch.squeeze(outputs).round().detach().numpy() == Y_train.detach().numpy())
        accuracy = 100 * correct/total
        losses.append(loss.item())
        accuracies.append(accuracy)
```

# With custom logistic regression using PyTorch without scaling

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using {device} device")
```

```python
def KLD_loss(y_true, y_pred):
    kld = nn.KLDivLoss(reduction="batchmean", log_target=False)
    loss = kld(y_pred, y_true) + kld(1-y_pred, 1-y_true)
    return loss
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self, loss_used):
        super(NeuralNetwork, self).__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(2, 1),
            nn.Sigmoid(),
        )
        if loss_used == "binary_cross_entropy":
            self.loss = nn.BCELoss(reduction="mean")
        else:
            self.loss = None
        self.learning_rate = 0.0001
        self.optimizer = torch.optim.SGD(self.parameters(),lr=self.learning_rate)
        self.losses = []
        self.accuracies = []

    def fit(self, x_train, y_train, epochs=10):
        for _ in range(epochs):
            logits = self.linear_relu_stack(x_train)
            if self.loss:
                loss = self.loss(logits, y_train.reshape(-1, 1))
            else:
                loss = KLD_loss(y_train.reshape(-1, 1), logits)
            self.losses.append(loss.detach().numpy())
            self.accuracies.append(accuracy_score(
                y_true=y_train, y_pred=logits.reshape(-1).detach().numpy().round()))
            loss.backward()
            self.optimizer.step()
        return logits
```
